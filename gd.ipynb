```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x * x - 2 * x + 1

def g(x):
    return 2 * x - 2

x = np.linspace(-5,7,100)
y = f(x)
```

```python
def gd(x_start, step, g):   # gd代表了Gradient Descent
    x = x_start
    for i in range(20):
        grad = g(x)
        x -= grad * step
        print '[ Epoch {0} ] grad = {1}, x = {2}'.format(i, grad, x)
        if abs(grad) < 1e-6:
            break;
    return x
gd(5,0.1,g)
```

```python
gd(5,100,g)
```

```python
gd(4,1,g)
```

```python
def f2(x):
    return 4 * x * x - 4 * x + 1
def g2(x):
    return 8 * x -4
gd(5,0.25,g2)
```
